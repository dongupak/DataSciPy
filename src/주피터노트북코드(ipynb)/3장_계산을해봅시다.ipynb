#### 컴퓨터과학에서 아주 중요한 나머지 연산자 : %

```python
p = int(input("분자를 입력하시오: "))    # 실수를 입력받을 때 int() 대신 float() 사용
q = int(input("분모를 입력하시오: "))    # 실수를 입력받을 때 int() 대신 float() 사용
print("나눗셈의 몫=", p // q)
print("나눗셈의 나머지=", p % q)
```

```python
number = int(input("정수를 입력하시오: "))
print(number%2)    # 나머지로 연산의 결과가 0이면 짝수, 1이면 홀수이다
```

### LAB 3-3 자동 판매기 프로그램을 만들어 보자

```python
money = int(input("투입한 돈: "))
price = int(input("물건값: "))

change = money - price
print("거스름돈: ", change)
coin500s = change // 500        # 500으로 나누어서 몫이 500원 동전의 개수
change = change % 500           # 500으로 나눈 나머지를 계산한다. 
coin100s = change // 100        # 100으로 나누어서 몫이 100원짜리의 개수

print("500원 동전의 개수:", coin500s)
print("100원 동전의 개수:", coin100s)
```

### 3.6 이진수를 잘 다루는 컴퓨터에 최적화된 연산자 : 비트 연산자

```python
bin(9)          # 2진수 형식 출력 (00001001)
```

```python
bin(10)         # 2진수 형식 출력 (00001010)
```

```python
bin(9 & 10)     # 9와 10을 2진수로 표현했을 때 모두 1인 위치만 1 나머지는 0이 된다
```

```python
bin(9 | 10)     # 9와 10을 2진수로 표현했을 때 하나로도 1이 나타나는 자리는 1이 된다
```

```python
bin(~9)         # 9의 모든 비트를 바꾸면 음수 10이 된다. 이것은 아래에 설명한다
```

```python
bin(9)          # 이진수 00001001
```

```python
bin(10)         # 이진수 00001010
```

```python
9 ^ 10          # 결과는 00000011 = 십진수 3
```

```python
bin(9 ^ 10)     # 앞 부분의 0은 모두 사라지고 1이 나타나는 곳부터 출력된다.
```

```python
4 << 1          # 00100을 한 비트 왼쪽으로 이동하여 01000(10진수 8을 만든다.
```

```python
4 << 2          # 00100을 두 비트 왼쪽으로 이동하여 10000(10진수 16)을 만든다.
```

```python
num = 2     # 정수 2는 이진수 10(2)으로 표현됨
num <<= 1   # 정수 2를 1비트 왼쪽 이동시키면 100(4)가 됨
num
```

```python
num <<= 1   # 정수 4를 1비트 더 왼쪽 이동시키면 1000(8)이 됨
num
```

```python
num >>= 1   # 정수 8를 1비트 오른쪽 이동시키면 100(4)가 됨
num
```

```python
num >>= 2   # 정수 4를 2비트 오른쪽 이동시키면 001(1)이 됨
num
```

### 3.8 random 모듈과 math 모듈로 다양한 기능을 사용해보자

```python
import random
random.random()             # 0 이상 1 미만의 임의의 실수를 반환
```

```python
random.random()             # 이 함수는 매번 다른 실수를 반환
```

```python
random.randint(1, 7)        # 1이상 7이하(7을 포함)의 임의의 정수를 반환
```

```python
random.randrange(7)         # 0이상 7미만(7을 포함하지 않음)의 임의의 정수를 반환
```

```python
random.randrange(1, 7)      # 1이상 7미만(7을 포함하지 않음)의 임의의 정수를 반환
```

```python
random.randrange(0, 10, 2)  # 0, 2, 4, 6, 8 중(10은 포함 안함) 하나를 반환함
```

```python
lst = [10, 20, 30, 40, 50]  # 여러 개의 값을 가지는 리스트를 생성함
random.shuffle(lst)         # lst 리스트의 순서를 무작위로 섞음
lst
```

```python
random.choice(lst)          # lst 리스트의 원소 중 무작위로 하나를 고름        
```
